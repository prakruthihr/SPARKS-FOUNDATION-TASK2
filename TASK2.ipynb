# **To find optimum number of cluster**

```python
from sklearn.cluster import KMeans
import pandas as pd


x=iris.iloc[:,1:4].values


cluster=[]
cluster_range=range(1,15)
for i in cluster_range:
  kmeans=KMeans(n_clusters=i,init='k-means++',n_init=10)
  kmeans.fit(x)
  cluster.append(kmeans.inertia_)
```

**ELBOW METHOD TO FIND OPTIMUM NUMBER OF CLUSTER**

```python
plt.plot(cluster_range,cluster,'go--',color='green')
plt.grid()
plt.title('ELBOW METHOD',fontsize=20)
plt.xlabel('NO OF CLUSTERS',fontsize=14)
plt.ylabel('within cluster sum of square',fontsize=14)
plt.show()
```

**APPLYING KMEANS CLUSTERING ON DATA**

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters =3,init='k-means++',max_iter=300,n_init=10,random_state=0)
model.fit(x)
prediction=model.predict(x)
```

## **visualising clusters**

```python
plt.figure(figsize=(10,6))
plt.scatter(x[prediction==0,0] ,x[prediction==0,1],s=25,c='indigo',label='iriis-setosa')
plt.scatter(x[prediction==1,0] ,x[prediction==1,1],s=25,c='blue',label='iriis-versicolor')
plt.scatter(x[prediction==2,0] ,x[prediction==2,1],s=25,c='green',label='iriis-verginica')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1], s=25,c='red',label='centroids',marker='*')
plt.grid()
plt.legend()
plt.show()
```
